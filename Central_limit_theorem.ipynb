**What is Central Limit Therorem(CLT)**

- Step-1:  If data does not follows normal distribution
    
- Step-2:  Take some samples has sample size n=20/30/100
    
- Step-3:  Contsruct means of each sample
    
- Step-4: plot the distribution plot on min 30 samples then it follows a normal distribution
    
    
This is called central limit theorem

# Method-2:

- In the above one we directly took 100 samples randomly

- No will take 100 samples randomly from a avialable data having 1000 observations

```python
import random
data=np.random.randint(1,1000,10000)

mean_s=[]
for i in range(100):
    s=random.sample(sorted(data),100)
    mean=np.mean(s)
    mean_s.append(mean)
```
